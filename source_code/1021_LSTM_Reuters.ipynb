```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import reuters
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
```

```python
(X_train, y_train), (X_test, y_test) = reuters.load_data(num_words=1000, test_split=0.2)
```

# 데이터 확인

```python
print(y_train)
category = np.max(y_train) + 1
print('카테고리 개수:', category)
```

```python
print('train 데이터 개수:', X_train.shape[0])
print('test 데이터 개수:', X_test.shape[0])
print(X_train[0])
```

```python
# 전체 단어 확인
word_index = reuters.get_word_index()
word_index
```

```python
index_to_word = {}
for key, value in word_index.items():
    index_to_word[value] = key
index_to_word
```

```python
' '.join([index_to_word[x] for x in X_train[0]])
```

```python
# 기사의 길이 확인
print('뉴스 기사의 최대 길이:', max(len(x) for x in X_train))
print('뉴스 기사의 최소 길이:', min(len(x) for x in X_train))
plt.plot([len(x) for x in X_train])
plt.show()
```

```python
plt.hist([len(x) for x in X_train], bins=50)
plt.show()
```

```python
print('기사 길이 평균:', sum(len(x) for x in X_train) / len(X_train))
```
